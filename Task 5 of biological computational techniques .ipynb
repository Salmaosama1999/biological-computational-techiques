## Human Digestion

```python
from pyopenms import *
from urllib.request import urlretrieve
```

```python
dig = ProteaseDigestion()
print(dig.getEnzymeName())
```

```python
seq = "".join([l.strip() for l in open("human_fasta.fasta").readlines()[1:]])
pro_seq = AASequence.fromString(seq)
print(pro_seq)
```

```python
degi_human=[]
dig.digest(pro_seq,degi_human)
print("The digestion of the sequence of the human gene is:")
for deige in degi_human:
    print(deige)
```

## Human Thoretical

```python
for s in degi_human :
    print('-'*50)
    print(s.toString())
    tsg = TheoreticalSpectrumGenerator()
    spec1 = MSSpectrum()
    p = Param()
    p.setValue("add_b_ions", "false")
    p.setValue("add_metainfo", "true")
    tsg.setParameters(p)
    tsg.getSpectrum(spec1,s, 1, 1)
    print("Spectrum of", s, "has", spec1.size(), "peaks.")
    for ion, peak in zip(spec1.getStringDataArrays()[0], spec1):
        print(ion.decode(), "is generated at m/z", peak.getMZ())    
```
